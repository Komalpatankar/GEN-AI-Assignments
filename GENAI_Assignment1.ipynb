```python
#Implement a basic probabilistic generative model (e.g., Gaussian Mixture Models).

import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
X = np.concatenate([
    np.random.normal(loc=[0, 0], scale=[1, 1], size=(100, 2)),
    np.random.normal(loc=[4, 4], scale=[1, 1], size=(100, 2)),
    np.random.normal(loc=[-4, 4], scale=[1, 1], size=(100, 2)),
])

# Fit a Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=0)
gmm.fit(X)

# Predict cluster assignments
labels = gmm.predict(X)

# Visualize the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=40, edgecolors='k')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.title("Gaussian Mixture Model Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
